```
import pickle
from keras.models import load_model
import numpy as np
import os
from google.colab.patches import cv2_imshow
import cv2
```

```
decoder = load_model('/content/drive/My Drive/Colab_Notebooks/Handwriting/decoder.model')
with open('/content/drive/My Drive/Colab_Notebooks/Handwriting/vectors.pickle', 'rb') as f:
    vectors = pickle.load(f)
```

```
def get_new(array):
  l= len(array)
  s = np.random.randint(low=0,high=l,size=1)[0]
  new_pred = array[s]
  for i in range(32):
    if int(new_pred[0][i])!=0:
      new_pred[0][i]+=np.random.uniform(low=0,high=0.1,size=1)[0]
  return np.array(new_pred).reshape(1,-1)
```

```
def generate_image(letter):
  new = get_new(vectors[letter])
  image = decoder.predict(new)[0,:,:,0]
  return image*255
datadir = '/content/dataset2/'
def get_image(letter):
  image_name = np.random.choice(os.listdir(datadir+letter))
  image = cv2.imread(datadir+letter+'/'+image_name,0)
  return image
```

```
cv2_imshow(generate_image('C'))
```

THIS METHOD FOR GENERATED CHARECTERS

```
def text_to_write(text):
  line = np.zeros((50,25))
  text = text
  for i in text:
    if i ==" ":
      line = np.append(line,np.zeros((50,25)),axis=1)
    elif i =="i":
      line = np.append(line[:,:-10],generate_image(i)[:,:-15],axis=1)
    else:
      line = np.append(line[:,:-10],generate_image(i),axis=1)
  return cv2.resize(line,(line.shape[1],50))
```

```
cv2_imshow(text_to_write('hello world'))
```

THIS METHOD IS FOR CONCATNATE EXISTING DATASET IMAGES 

```
def text_to_concat(text):
  line = np.zeros((50,25))
  text = text
  for i in text:
    if i ==" ":
      line = np.append(line,np.zeros((50,25)),axis=1)
    elif i in ["i","1"]:
      line = np.append(line[:,:-10],get_image(i)[:,:-15],axis=1)
    else:
      line = np.append(line[:,:-10],get_image(i),axis=1)
  return cv2.resize(line,(line.shape[1],50))
```

```
cv2_imshow(text_to_concat('hello world'))
```
