```
def binary(x_axis,rejection=0.1):
  flag = 0
  starts=[]
  ends = []
  for i in range(len(x_axis)-1):
    if x_axis[i]==0 and x_axis[i+1]==1 and flag==0:
      flag=1
      starts.append(i)
    if x_axis[i]==1 and x_axis[i+1]==0 and flag==1 :
      flag=0
      ends.append(i)
  return list(zip(starts,ends))
def binary_single(x_axis,rejection=0.1):
  #plt.plot(x_axis)
  #plt.show()
  flag = 0
  starts=[]
  ends = []
  for i in range(len(x_axis)-1):
    if x_axis[i]==0 and x_axis[i+1]==1:
      flag=1
      starts.append(i)
    if x_axis[i]==1 and x_axis[i+1]==0:
      flag=0
      ends.append(i)
  if len(starts)==0:
    starts.append(0)
  if len(ends)==0:
    ends.append(len(x_axis)-1)
  return list(zip(starts,ends))
def axis_cov(img,axis=0):
  y_axis = np.sum(img,axis=axis)
  y_axis = y_axis/np.max(y_axis)
  return y_axis
```

```
x_co = binary(np.array(x_axis>0.1,np.int))
y_co = binary(np.array(y_axis>0.11,np.int))
```
