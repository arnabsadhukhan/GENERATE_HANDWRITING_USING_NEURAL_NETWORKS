```
text = list('0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ(){}[]+-=:%')+['slash']
datadir = '/content/dataset/'
```

```
def binary_single(x_axis,rejection=0.05):
  #plt.plot(x_axis)
  #plt.show()
  flag = 0
  starts=[]
  ends = []
  for i in range(len(x_axis)-1):
    if x_axis[i]==0 and x_axis[i+1]==1:
      flag=1
      starts.append(i)
    if x_axis[i]==1 and x_axis[i+1]==0:
      flag=0
      ends.append(i)
  if len(starts)==0:
    starts.append(0)
  if len(ends)==0:
    ends.append(len(x_axis)-1)
  s,e = list(zip(starts,ends))[0]
  if e-s<len(x_axis)/3:
    return [(starts[0],len(x_axis)-1)]
  return list(zip(starts,ends))
```

```
try:
    os.mkdir('/content/dataset2')
except:pass
```

```
for folder in text:
  try:
      os.mkdir('/content/dataset2/'+folder)
  except:pass
  image_no=0
  for img in os.listdir(datadir+folder):
    print(img)
    image = cv2.imread(datadir+folder+'/'+img,0) 

    y_axis = np.sum(image,axis=1)/np.max(np.sum(image,axis=1))
    w,h=binary_single(y_axis>0.01)[0]

    x_axis = np.sum(image,axis=0)/np.max(np.sum(image,axis=0))
    x,y=binary_single(x_axis>0.01)[0]

    image = image[max(0,w-4):h+4,max(0,x-4):y+4]

    h,w =image.shape

    s=5

    pad = np.zeros((h+(s*2),w+(s*2)))

    pad[s:s+h,s:w+s] = image
    pad= cv2.resize(pad,(50,50))
    cv2_imshow(pad)
    cv2.imwrite('/content/dataset2/'+folder+'/'+str(image_no)+'.png',pad)
    image_no+=1


```
